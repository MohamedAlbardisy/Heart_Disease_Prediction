#### Logistic Regression model

```python
data = pd.read_csv('heart.csv')
data.head()
```

```python
data['Sex'] = data.Sex.replace({'M' : 1 , 'F' : 0})
```

```python
data['ChestPainType'] = data.ChestPainType.replace({'ASY' : 0 , 'NAP' : 1 , 'ATA' : 2 , 'TA' : 3})
```

```python
data['RestingECG'] = data.RestingECG.replace({'ST' : 0 , 'LVH' : 1 , 'Normal' : 2 })
```

```python
data['ExerciseAngina'] = data.ExerciseAngina.replace({'N' : 0 , 'Y' : 1 })
```

```python
data.ST_Slope = data.ST_Slope.replace({'Flat' : 0 , 'Up' : 1 , 'Down' : 2})
```

```python
x = data.drop('HeartDisease' , axis=1)
y = data.HeartDisease
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x , y , train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain , ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
y_predicted = model.predict(xtest)
```

```python
mean_absolute_error(ytest , y_predicted)
```
